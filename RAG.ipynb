```python
import os
import fitz  # PyMuPDF
import faiss
import numpy as np
import torch
from google.colab import files
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_openai import ChatOpenAI
from langchain.schema import Document
from langchain.chains import RetrievalQA

# Set OpenRouter API key and endpoint
os.environ["OPENROUTER_API_KEY"] = "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"
os.environ["OPENAI_API_KEY"] = os.environ["OPENROUTER_API_KEY"]
os.environ["OPENAI_API_BASE"] = "https://openrouter.ai/api/v1"

# Initialize LLM from OpenRouter
llm = ChatOpenAI(
    model="openai/gpt-3.5-turbo",
    openai_api_base=os.environ["OPENAI_API_BASE"],
    api_key=os.environ["OPENAI_API_KEY"]
)

# Function to upload PDF
def upload_pdf():
    print("\n📂 Please upload a PDF file.")
    uploaded = files.upload()
    pdf_path = list(uploaded.keys())[0]
    return pdf_path

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = "\n".join([page.get_text("text") for page in doc])
    return text

# Create FAISS vector store
def create_vector_store(text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    text_chunks = text_splitter.split_text(text)
    documents = [Document(page_content=chunk) for chunk in text_chunks]

    embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vector_store = FAISS.from_documents(documents, embedding_model)
    return vector_store

# Query answer function
def answer_query(query, qa_chain):
    response = qa_chain.invoke({"query": query})
    return response["result"]

# === MAIN LOOP ===
run_app = True
while run_app:
    pdf_path = upload_pdf()
    document_text = extract_text_from_pdf(pdf_path)
    vector_store = create_vector_store(document_text)
    retriever = vector_store.as_retriever()
    qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)

    while True:
        query = input("\n🔍 Enter your query (or type 'upload' to upload a new PDF, 'exit' to quit): ").strip()

        if query.lower() == "upload":
            break  # Go back to outer loop

        if query.lower() == "exit":
            print("\n👋 Exiting. Have a great day!")
            run_app = False
            break  # Break inner loop too

        print("\n💡 Answer:\n", answer_query(query, qa_chain))
```
