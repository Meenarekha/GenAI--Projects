Sentiment Analysis


```python
import gradio as gr
from transformers import pipeline

sentiment =pipeline("sentiment-analysis")

def analyze_sentiment(text):
    result = sentiment(text)[0]
    return result['label'], result['score']

interface=gr.Interface(fn=analyze_sentiment, inputs='text',outputs='json',title='Sentiment Analysis')

interface.launch()
```

Text Generation

```python
import gradio as gr
from transformers import pipeline

# Load the text generation model (GPT-2)
generator = pipeline("text-generation", model="gpt2")

# Define the function for generating text
def generate_text(prompt):
    return generator(prompt, max_length=100, num_return_sequences=1)[0]['generated_text']


inputs = gr.Textbox(lines=3, placeholder="Type your prompt here...", label="Enter Prompt")
interface = gr.Interface(fn=generate_text, inputs=inputs, outputs="text", title="Text Generation")

# Launch the interface
interface.launch()
```

Text Summarization

```python
import gradio as gr
from transformers import pipeline

# Load the summarization model
summarizer = pipeline("summarization")

# Define the function for summarizing text
def summarize_text(text):
    return summarizer(text, max_length=150, min_length=50, do_sample=False)[0]['summary_text']

# Create the Gradio interface
inputs = gr.Textbox(lines=3, placeholder="Type your prompt here...", label="Enter Prompt")
interface = gr.Interface(fn=summarize_text, inputs=inputs, outputs="text",title="Text Summarization")

# Launch the interface
interface.launch()
```

Text Translation

```python
import gradio as gr
from transformers import pipeline

#english to french
translator = pipeline("translation_en_to_fr")

def translate_text(text):
    return translator(text)[0]['translation_text']

inputs = gr.Textbox(lines=3, placeholder="Type your prompt here...", label="Enter Prompt")
interface = gr.Interface(fn=translate_text,inputs=inputs,outputs="text",title="Text Translation")

interface.launch()
```

```python
import gradio as gr
from transformers import pipeline

# Define the translation function for different languages with explicit models
def translate_text(text, target_language):
    try:
        # Specify the model explicitly for each language pair
        if target_language == "French":
            translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")
        elif target_language == "Spanish":
            translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-es")
        elif target_language == "German":
            translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-de")
        elif target_language == "Italian":
            translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-it")
        elif target_language == "Dutch":
            translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-nl")
        elif target_language == "Russian":
            translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ru")
        else:
            return f"Sorry, the language '{target_language}' is not supported."

        # Perform translation
        translation = translator(text)[0]['translation_text']
        return translation

    except Exception as e:
        return f"An error occurred: {str(e)}"

# Create Gradio interface with a dropdown for selecting language
language_dropdown = gr.Dropdown(
    choices=[
        "French", "Spanish", "German", "Italian",
        "Dutch", "Russian"
    ],
    label="Select Language"
)

inputs = gr.Textbox(lines=5, placeholder="Type your text here...", label="Enter Text")

# Create the Gradio Interface
interface = gr.Interface(
    fn=translate_text,
    inputs=[inputs, language_dropdown],
    outputs="text",
    title="Multi-language Translator"
)

# Launch the interface
interface.launch()
```

```python
import gradio as gr
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

def pos_tagging(text):
    doc = nlp(text)
    tagged_words = [(token.text, token.pos_) for token in doc]
    return tagged_words

iface = gr.Interface(
    fn=pos_tagging,
    inputs=gr.Textbox(lines=2, placeholder="Enter text here..."),
    outputs=gr.Dataframe(headers=["Word", "Part of Speech"], label="POS Tags"),
    title="Parts of Speech Tagging",
    description="Enter a sentence to see its parts of speech tagging."
)

iface.launch()
```

```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load grammar correction model
model_name = "prithivida/grammar_error_correcter_v1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
grammar_corrector = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

# Define correction function
def correct_grammar(text):
    correction = grammar_corrector(text, max_length=128, clean_up_tokenization_spaces=True)[0]['generated_text']
    return correction

# Gradio interface
iface = gr.Interface(
    fn=correct_grammar,
    inputs=gr.Textbox(lines=3, placeholder="Enter sentence with grammar/spelling errors..."),
    outputs="text",
    title="Grammar & Spell Checker",
    description="Corrects grammar and spelling errors in English text."
)

iface.launch()
```
